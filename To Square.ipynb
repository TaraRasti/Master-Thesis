```python
def row(x, inx):
    x = np.insert(x, inx, np.zeros((image.shape[1], 1)), axis = 0)
    return x

def col(x, inx):
    x = np.insert(x, inx, np.zeros((x.shape[0], 1)), axis = 1)
    return x
```

```python
for k in range(len(path1)):
    from matplotlib import image
    filepath = path1[k]
    image  = image.imread(filepath)
    if (image.shape[0]<256):
        x = row(image, image.shape[0])
        for i in range(image.shape[0] + 1,255):
            x = row(x, i)
        for j in range(x.shape[1], 255):
            x = col(x, j)         
    file_path =  '/Users/taraapple/Downloads/Demented1/' + L2[k]
    plt.imsave(file_path, x, format='jpg', cmap = 'gray')                       
```
